```python
# Absolute Maximum Scaling

import numpy as np
# Original data
data = np.array([[-50], [0], [30], [70], [-90]])
# Absolute Maximum Scaling
scaled_data = data / np.max(np.abs(data))
print("Original Data:\n", data)
print("Scaled Data:\n", scaled_data)
```

```python
# Min-Max Scaling

import numpy as np
from sklearn.preprocessing import MinMaxScaler
# Sample data
data = np.array([[100], [200], [300], [400], [500]])
# Min-Max Scaling
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
print("Original Data:\n", data)
print("Scaled Data:\n", scaled_data)
```

```python
# Normalization

import numpy as np
from sklearn.preprocessing import Normalizer
# Sample dataset (2 features per sample)
data = np.array([[3, 4],
               [1, 2],
               [2, 0]])
# Apply L2 Normalization
normalizer = Normalizer(norm='l2')
normalized_data = normalizer.transform(data)
print("Original Data:\n", data)
print("Normalized Data:\n", normalized_data)
```

```python
# Standardization

import numpy as np
from sklearn.preprocessing import StandardScaler
# Sample data
data = np.array([[10], [20], [30], [40], [50]])
# Apply Standardization
scaler = StandardScaler()
standardized_data = scaler.fit_transform(data)
print("Original Data:\n", data)
print("Standardized Data:\n", standardized_data)
```

```python
# Assignment 2: Heart CSV: try to identify categorical/discrete data and encode it.
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder


file_path = '/content/drive/MyDrive/mt/heart.csv'
#dataset = pd.read_csv(file_path)

df = pd.read_csv(file_path)

# Display first 5 rows
print(df.head())


# Step 1: Inspect the dataset
print("\n\nColumns:\n", df.columns)
print(df)



# Step 2: Identify categorical/discrete columns
# Typically, columns like 'sex', 'cp' (chest pain type), 'fbs', 'restecg', 'exang', 'slope', 'thal' are categorical
categorical_cols = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'thal']


print('Encode categorical columns\n')
# Step 3: Encode categorical columns
# Option A: Label Encoding (good for ordinal or when model can handle integers)

# Create a copy for encoded data
df_encoded = df.copy()

# Apply Label Encoding

label_encoder = LabelEncoder()
for col in categorical_cols:
    print(f"Encoding column: {col}")
    print("Before:", df_encoded[col].unique())  # Check original values
    df_encoded[col] = label_encoder.fit_transform(df_encoded[col])
    print("After:", df_encoded[col].unique())   # Check transformed values
    print("-" * 40)


# Print the new DataFrame
print(df_encoded.head())


```

```python
# Assignment 3: Weather Aus: Find out columns with missing values (NA) and identify the ratio of missing values.
# Apply mean, median and mode imputation based on the column type. Observe the changes after imputation.
# Use simple plot.
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns



file_path = '/content/drive/MyDrive/mt/weatherAUS.csv'
df = pd.read_csv(file_path)


# Step 1: Find columns with missing values and calculate missing ratio
print('\nFind columns with missing values and calculate missing ratio\n');
missing_counts = df.isnull().sum()
missing_ratio = (missing_counts / len(df)) * 100
missing_info = pd.DataFrame({'MissingCount': missing_counts, 'MissingRatio(%)': missing_ratio})
print(missing_info)


# Filter columns with missing values
print('\nFilter columns with missing values\n')
missing_info = missing_info[missing_info['MissingCount'] > 0]
print(missing_info)


plt.figure(figsize=(8, 5))
sns.barplot(x=missing_ratio.index, y=missing_ratio.values)
plt.xticks(rotation=45)
plt.title("Missing Value Ratio Before Imputation")
plt.ylabel("Ratio")
plt.tight_layout()
plt.savefig('missing_ratio_before.png')
plt.show()


# Step 2: Apply mean, median, and mode imputation based on column type
print('\nApply mean, median, and mode imputation based on column type\n')
df_imputed = df.copy()

for column in df_imputed.columns:
    if df_imputed[column].isnull().sum() > 0:
        if df_imputed[column].dtype in ['float64', 'int64']:
            # For numeric columns, use mean imputation
            mean_value = df_imputed[column].mean()
            df_imputed[column] = df_imputed[column].fillna(mean_value)
        else:
            # For categorical columns, use mode imputation
            mode_value = df_imputed[column].mode()[0]
            df_imputed[column] = df_imputed[column].fillna(mode_value)
print(df_imputed)



# Step 3: Observe changes after imputation
print('\nObserve changes after imputation\n')
missing_counts_after = df_imputed.isnull().sum()
missing_ratio_after = (missing_counts_after / len(df_imputed)) * 100
missing_info_after = pd.DataFrame({'MissingCount': missing_counts_after, 'MissingRatio(%)': missing_ratio_after})
missing_info_after = missing_info_after[missing_info_after['MissingCount'] > 0]
print(missing_info_after)



# Step 4: Visualize missing values before and after imputation using a simple bar plot
print('\nVisualize missing values before and after imputation using a simple bar plot\n');


# Plot missing value ratio after imputation
plt.figure(figsize=(8, 5))
sns.barplot(x=missing_ratio.index, y=[0]*len(missing_ratio))  # All zeros after imputation
plt.xticks(rotation=45)
plt.title("Missing Value Ratio After Imputation")
plt.ylabel("Ratio")
plt.tight_layout()
plt.savefig("missing_ratio_after.png")
plt.show()


```
