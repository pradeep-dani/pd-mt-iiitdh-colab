```python
# Step 1: Get our drawing tools
import matplotlib.pyplot as plt
```

```python
# Step 2: Our data (like plotting points on graph paper)
days = [1, 2, 3, 4, 5]  # Monday to Friday
happiness = [3, 4, 5, 4, 5]  # Happiness level (1-5)
```

```python
# Step 3: Create the line chart
plt.plot(days, happiness)
```

```python
# Step 3: Add labels (so people know what they're looking at!)
plt.xlabel('Day of the Week')
plt.ylabel('Happiness Level')
plt.title('My Week in Happiness')
# Step 4: Create the line chart
plt.plot(days, happiness)
# Step 5: Show the chart
plt.show()
```

```python
#Example: Plotting Temperature
hours = [6, 9, 12, 15, 18, 21]  # Time of day
temperature = [15, 18, 25, 28, 22, 17]  # Temperature in Celsius
plt.plot(hours, temperature, marker='o')  # 'o' adds dots!
plt.xlabel('Hour of Day')
plt.ylabel('Temperature (C)')
plt.title('Temperature Throughout the Day')
plt.show()
```

```python
# Example: Favorite Ice Cream Flavors
# Our data
flavors = ['Vanilla', 'Chocolate', 'Strawberry', 'Mint']
votes = [15, 25, 12, 8]
# Create colorful bars
colors = ['#FFE5B4', '#7B3F00', '#FFB6C1', '#98FB98']
plt.bar(flavors, votes, color=colors)
plt.xlabel('Ice Cream Flavors')
plt.ylabel('Number of Votes')
plt.title('Class Favorite Ice Cream Poll')
plt.show()
```

```python
# Example: How You Spend Your Day
# Daily activities
activities = ['Sleep', 'Work', 'Meals', 'Exercise', 'Leisure', 'Commute']
hours = [8, 8, 2, 1, 4, 1]
# Create a pie chart with colors
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFD700', '#FF99CC', '#C0C0C0']
plt.pie(hours, labels=activities, colors=colors, autopct='%1.1f%%')
plt.title('How I Spend My 24 Hours')
plt.show()
```

```python
#Example: Study Time vs Test Scores
# Student data
study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
test_scores = [50, 55, 65, 70, 75, 82, 88, 95]
# Create scatter plot
plt.scatter(study_hours, test_scores, color='blue', s=100)
plt.xlabel('Hours Studied')
plt.ylabel('Test Score')
plt.title('Does Studying More Help?')
plt.grid(True, alpha=0.3)  # Add a light grid
plt.show()
```

```python
#Creating a Professional Bar Plot
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data - Monthly sales
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [45000, 52000, 48000, 61000, 58000]

# Seaborn's stylish bar plot
sns.barplot(x=months, y=sales, palette='viridis')
plt.title('Monthly Sales Performance')
plt.ylabel('Sales ()')
plt.show()
```

```python
#Example: Weekly Exercise Tracker
import numpy as np
# Exercise minutes per day
exercise_data = [
    [30, 45, 0, 60, 30, 90, 0],   # Week 1
    [45, 30, 30, 45, 0, 120, 60],  # Week 2
    [60, 45, 30, 30, 45, 90, 0],   # Week 3
    [30, 60, 45, 0, 30, 105, 45]   # Week 4
]
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
weeks = ['Week 1', 'Week 2', 'Week 3', 'Week 4']
# Create heatmap
sns.heatmap(exercise_data, xticklabels=days, yticklabels=weeks,
            cmap='YlOrRd', annot=True)
plt.title('Monthly Exercise Tracker (minutes)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
np.random.seed(42)  # For reproducibility
x = np.arange(1, 13)  # Months 1-12
y = np.random.randint(40, 90, 12)  # Random revenue between 40-90K

# Month labels
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Create figure with custom size
plt.figure(figsize=(10, 6))

# Plot with style
plt.plot(x, y,
         color='#2E86AB',      # Custom color
         linewidth=2.5,        # Thicker line
         marker='o',           # Add markers
         markersize=8,         # Bigger markers
         label='Sales 2024')   # Add legend label

# Set month labels
plt.xticks(x, months, rotation=45)  # Rotate labels if needed

# Professional formatting
plt.xlabel('Month', fontsize=12, fontweight='bold')
plt.ylabel('Revenue (K)', fontsize=12, fontweight='bold')
plt.title('2024 Sales Performance', fontsize=16, pad=20)
plt.legend(loc='best')
plt.grid(True, alpha=0.3)
plt.tight_layout()  # Prevents label cutoff
plt.show()

# Print the data to see what was plotted
print("Monthly Revenue Data (K):")
for month, revenue in zip(months, y):
    print(f"{month}: {revenue}K")
```

## *Your Mission:*
Visualize "A Week of Weather" with this data:

Days: Monday to Sunday  
Temperature: [22, 24, 19, 18, 25, 27, 23]C  
Rainfall: [0, 5, 12, 8, 0, 0, 3]

```python

import matplotlib.pyplot as plt
import seaborn as sns
# Set style for beautiful plots
sns.set_style("whitegrid")
# Our data
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperature = [22, 24, 19, 18, 25, 27, 23]
rainfall = [0, 5, 12, 8, 0, 0, 3]
# Create subplots
xfig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))
# Temperature plot (top)
ax1.plot(days, temperature, color='#FF6B6B', marker='o', linewidth=2, markersize=8)
ax1.fill_between(range(len(days)), temperature, alpha=0.3, color='#FF6B6B')
ax1.set_title('Weekly Temperature Trend', fontsize=14, fontweight='bold')
ax1.set_ylabel('Temperature (C)')
ax1.grid(True, alpha=0.3)
# Rainfall plot (bottom)
ax2.bar(days, rainfall, color='#4ECDC4', alpha=0.7, edgecolor='#45B7B8', linewidth=2)
ax2.set_title('Weekly Rainfall', fontsize=14, fontweight='bold')
ax2.set_ylabel('Rainfall (mm)')
ax2.set_xlabel('Day of Week')
# Add value labels on bars
for i, v in enumerate(rainfall):
  ax2.text(i, v + 0.3, f'{v}mm', ha='center', fontsize=9)
  plt.tight_layout()
plt.show()
```
